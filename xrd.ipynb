```python
import matplotlib.pyplot as plt
from pymatgen.analysis.diffraction.xrd import *
from pymatgen.analysis.diffraction.core import *
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester ("hGWi2mG3AMC2JfpV0r") as m:
    
   
    structure = m.get_structure_by_material_id("mp-36111")
    ConventionalStructure = SpacegroupAnalyzer(structure).get_conventional_standard_structure()
    xrd = XRDCalculator(1.54060) 
    pattern = xrd.get_pattern(ConventionalStructure) 
```

```python
xrd.show_plot(ConventionalStructure)
print("Voici le graphique reprsentant les pics du diffractogramme pour une longueure d'onde =1.54060  ")
```

```python
angle = pattern.x #Tableau contenant les angles 2*theta
hkl = pattern.hkls
print("\n Le premier pic sera observ avec un angle 2 =",angle[0]," et correspond au plan",hkl[0][0]["hkl"])
print("\n Le second pic sera observ avec un angle 2 =",angle[1]," et correspond au plan",hkl[1][0]["hkl"])
print("\n Le troisime pic sera observ avec un angle 2 =",angle[2]," et correspond au plan",hkl[2][0]["hkl"])

```
